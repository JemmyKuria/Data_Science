```python
# Calculate Median of the list below

# List of Numbers
nums = [12, 10, 13, 8, 5 , 6, 7, 8, 21, 17, 20, 19]

# Sorting List
sorted_nums = sorted(nums)
print(sorted_nums)

# Find Middle Element
num_len = len(sorted_nums)

# a = num_len / 2 # Returns the division value
# a1 = num_len % 2 # Returns the remainder

if num_len % 2 == 1:
    median = (num_len - 1) // 2
    print(sorted_nums[median])

else:
    median = num_len // 2
    median2 = median - 1
    average = (sorted_nums[median] + sorted_nums[median2])/2
    print(average)
```

```python
#create a sample dataset
np.random.seed(10) #ensures reproducibility. Same results each time it runs

data ={
    "Month":["Jan", "Feb", "March", "April", "May", "Jun"],
    "Sales": np.random.randint(200, 500, 6),
    "Customers":np.random.randint(50, 200, 6),
    "Profit": np.random.randint(50, 150, 6)
}
data
```

```python
#create a sample dataset
np.random.seed(5) #ensures reproducibility. Same results each time it runs

data2 ={
    "Month":["Jan", "Feb", "March", "April", "May", "Jun"],
    "Sales": np.random.randint(100, 400, 6),
    "Customers":np.random.randint(100, 250, 6),
    "Profit": np.random.randint(100, 200, 6)
}
data2
```

```python
df=pd.DataFrame(data)
df
```

```python
df2=pd.DataFrame(data2)
df2
```

```python
plt.figure(
    figsize=(20, 7)) #shows the size of the plot, width and height
plt.plot(df["Month"], #x axis
         df["Sales"], #y axis
         color="r", # color of the line
         marker="o", # marker of the data points
         linestyle="-", #connects the line using
         label="Sales")
plt.plot(df2["Month"], #x axis
         df2["Sales"], #y axis
         color="g", # color of the line
         marker="o", # marker of the data points
         linestyle="-", #connects the line using
         label="Sales2")
plt.title("Sales vs Month")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()#
plt.grid(True)
plt.show()


```
