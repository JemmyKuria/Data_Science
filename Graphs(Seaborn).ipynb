```python
#Histogram. Visual that shows the distribution of values in a dataset. Made up of bars and frequencies.
#x axis= bins and y axis= frequencies



#Scatter plot. Visual that shows the relationship between 2 variables. Each point reps an observation. X shows one variable and Y shows the other.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Load an inbuilt dataset
data=sns.load_dataset("tips")
data.head()
```

```python
data.info()
```

```python
#Histogram
plt.figure(figsize=(10, 7))
sns.histplot(data['total_bill'], bins=10, palette="coolwarm", hue="total_bill")
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
#Box plot
#Box plot. Visual that summarizes data in 5 categories: minimum(excepts outliers), first quartile(25th percentile), median(q2),
#third quartile(75th percentile), maximum(except outliers)
#It also includes whiskers(lines extending from the box) and also possible outliers(plotted as dots outside the whiskers)
plt.figure(figsize=(10, 7))
sns.boxplot(data=data, x="day", y="tip" , palette="coolwarm", hue="day")
plt.title("Tips By Day")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.regplot(x="total_bill", y="tip", data=data, line_kws={"color": "red"} )
plt.title("Total Bill vs Sex")
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()
```
