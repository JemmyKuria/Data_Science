```python
# List of numbers
x = [3, 5, 8, 12, 15, 18, 20, 22, 25, 30, 50, 80, 687]
sorted_list=sorted(x)
print(sorted_list)
diff=max(sorted_list)-min(sorted_list)
print(diff)
```

```python
# Sort in ascending order
x = sorted(x)
```

```python
# Distance between last and first element
distance = len(x) - 1
```

```python
# Multiply distance by percentiles

# Index of 25th percentile
index_p25 = 0.25*distance
index_p25
```

```python
# Index of 75th percentile
index_p75 = 0.75*distance
index_p75
```

```python
# 25th Percentile
p25 = x[int(index_p25)]
p25
```

```python
# 75th Percentile
p75 = x[int(index_p75)]
p75
```

```python
# IQR
iqr = p75 - p25
iqr
```

```python
import numpy as np

np.percentile(x, 75) - np.percentile(x, 25)
```
